```python
import pandas as pd
bookings = pd.read_csv(r'C:\GeekBrains\ETL\HW\8\booking.csv')
clients = pd.read_csv(r'C:\GeekBrains\ETL\HW\8\client.csv')
hotels = pd.read_csv(r'C:\C:\GeekBrains\ETL\HW\8\hotel.csv')
import json

print(bookings)
```

```python
data = pd.merge(bookings, clients, on='client_id')
data.rename(columns={'name': 'client_name', 'type': 'client_type'}, inplace=True)
data
```

```python
data = pd.merge(data, hotels, on='hotel_id')
data.rename(columns={'name': 'hotel_name'}, inplace=True)
data
```

```python
data['booking_date'] = data['booking_date'].apply(lambda x: pd.to_datetime(x).strftime('%Y-%m-%d'))
data
```

```python

data_dict = json.loads(data.to_json())

data_dict
```

```python
all_data = pd.DataFrame(data_dict)
all_data
```

```python
all_data['age'].fillna(all_data['age'].mean(), inplace = True)
all_data['age'] = all_data['age'].astype(int)
all_data.booking_cost.fillna(all_data[all_data.room_type == 'standard_1_bed'].booking_cost.mean(), inplace=True)
```

```python
all_data.loc[all_data.room_type == 'standard_1_bed', ['currency']] = 'GBP'
all_data
```

```python
all_data.loc[all_data.currency == 'EUR', ['booking_cost']] = all_data.booking_cost * 0.85
all_data.currency.replace("EUR", "GBP", inplace=True)
all_data
```
